```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#  Load raw data into 'games' dataframe
games = pd.read_csv("./datasets/steam-games.csv")

#  Remove columns that are unnecessary to the project
games.drop(columns=['release_date', 'app_id', 'developer',
                    'publisher', 'age_rating', 'original_price', 'discount_percentage',
                    'discounted_price', 'dlc_available', 'content_descriptor', 'about_description', 'awards',
                    'recent_review', 'recent_review_%', 'recent_review_count'], inplace=True)
games.info()
```

```python
# Histogram for overall_review_%
plt.figure(figsize=(10, 5))
plt.hist(games['overall_review_%'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Overall Review Percentage')
plt.xlabel('Overall Review %')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Examine categories
categories = games['categories'].explode()
categories
```

```python
#  Removing some irrelevant categories like 'Steam Achievements', 'Steam Leaderboards'
unwanted_cat = ['Family Sharing', 'Steam Achievements', 'Profile Features Limited', 'Steam Cloud',
                'Steam Trading Cards', 'Steam Leaderboards', 'Remote Play Together', 'Tracked Controller Support',
                'Stats', 'VR Only', 'Remote Play on TV', 'In-App Purchases',
                'Steam Workshop', 'Includes level editor', 'Steam is learning about this game',
                'Captions available', 'Remote Play on Tablet', 'MMO', 'Remote Play on Phone']
for cat in unwanted_cat:
    games['categories'] = games['categories'].str.replace(cat, '')

#  Categories and genres contain various string values seperated by commas which can be 
#  cleanned (delete duplicates) and make into columns
games['categories'] = games['categories'].str.strip(',')

#  apply(lambda x: x if isinstance(x, list) else []) => Replace any non-list values (such as NaN) with empty lists.
games['categories'] = games['categories'].str.split(', ').apply(lambda x: x if isinstance(x, list) else [])
```

```python
# Select 10 most occurring categories
categories = games['categories'].explode()
top_categories = categories.value_counts().nlargest(10).index.tolist()
top_categories.pop(0) # Top categories is '' (empty)
top_categories
```

```python
# Create columns with categories names
for category in top_categories:
    games[f'{category}'] = games['categories'].apply(lambda x: category in x)
games.drop(columns=['categories'], inplace=True)

games.info()
```

```python
# Create 'LAN Multiplayer' column by combining 
games['LAN Multiplayer'] = (games['Shared/Split Screen PvP'] | 
                                     games['Shared/Split Screen Co-op'] | 
                                     games['LAN PvP'] | 
                                     games['LAN Co-op'])

# Create 'Online Multiplayer' column
games['Online Multiplayer'] = (games['Online Co-op'] | games['Online PvP'])
games.drop(columns=['Shared/Split Screen PvP', 'Shared/Split Screen Co-op', 
                    'LAN PvP', 'LAN Co-op',
                    'Online Co-op', 'Online PvP'], inplace=True)
games.info()
```

```python
# Sum of boolean features for plotting
features = games.columns[8:]
feature_counts = games[features].sum()

# Bar plot for feature prevalence
plt.figure(figsize=(14, 8))
feature_counts.plot(kind='bar')
plt.title('Game Features')
plt.xlabel('Feature')
plt.ylabel('Number of Games')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Same process on genres, but we select top 6 only.
# apply(lambda x: x if isinstance(x, list) else []) => Replace any non-list values (such as NaN) with empty lists.
games['genres'] = games['genres'].str.split(', ').apply(lambda x: x if isinstance(x, list) else [])

genres = games['genres'].explode()
top_genres = genres.value_counts().nlargest(13).index.tolist()
top_genres
```

```python
# Create columns with genres names
for genre in top_genres:
    games[f'{genre}'] = games['genres'].apply(lambda x: genre in x)
games.drop(columns=['genres'], inplace=True)

games.drop(columns=['Early Access', 'Education'], inplace=True)
```

```python
# Sum of boolean features for plotting
features = games.columns[12:]
feature_counts = games[features].sum()

# Bar plot for feature prevalence
plt.figure(figsize=(14, 8))
feature_counts.plot(kind='bar')
plt.title('Game Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
########################################################## Review Mapping #########################################################################
# Quantifying reviews
review_score = {'Overwhelmingly Positive': 10,
                'Very Positive': 9,
                'Mostly Positive': 8,
                'Positive': 7,
                'Mixed': 6,
                'Negative': 5,
                'Mostly Negative': 4,
                'Very Negative': 3,
                'Overwhelmingly Negative': 3}
games['overall_review'] = games['overall_review'].map(review_score).fillna(3)
```

```python
import seaborn as sns

# Histogram of overall reviews
plt.figure(figsize=(10, 6))
sns.histplot(games['overall_review'], bins=20, kde=True)
plt.title('Distribution of Overall Review Scores')
plt.xlabel('Overall Review Score')
plt.ylabel('Frequency')
plt.show()
```

```python
data = games[games['overall_review_count'] < 10000].copy()

# Convert Name to a numerical index
data['Name_index'] = data['title'].astype('category').cat.codes

# Create the scatter plot
plt.figure(figsize=(14, 8))
plt.scatter(data['Name_index'], data['overall_review_count'], alpha=0.6)
# plt.xticks(m_games['Name_index'], m_games['Name'], rotation='vertical', fontsize=8)
plt.xlabel('Game Name')
plt.ylabel('Overall Review %')
plt.title('Scatter Plot of Game Names vs. Overall Review %')
plt.grid(True)
plt.tight_layout()
plt.show()
```
